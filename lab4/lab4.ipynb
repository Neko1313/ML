```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("../db/df_no_duplicates.json")
df
```

```python
df.columns
```

```python
df_new = df[["manga_name", '10', '9', '8','7', '6', '5', '4', '3', '1', '2']]
df_new
```

```python
df_new = df_new.fillna(0)
df_new = df_new.set_index('manga_name')
```

```python
df_new
```

```python
df_new = df_new[["10", "1"]]
```

```python
df_test = df_new
```

```python
df_test["10"] = df_test["10"]/100000
```

```python
df_test["1"] = df_test["1"]/10000
```

```python
df_test
```

```python
df_test = df_test[df_test["10"] > 0.001]
```

```python
df_test
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
x_scaler = StandardScaler().fit_transform(df_test)

# Инициализация и обучение модели DBSCAN
dbscan = DBSCAN(eps=0.35, min_samples=2)
dbscan.fit(x_scaler)

labels = dbscan.labels_

# Визуализация результатов, темные точки - выбросы
plt.scatter(df_test.to_numpy()[:, 0], df_test.to_numpy()[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
set(dbscan.labels_)
```

```python
df_test['labels'] = dbscan.labels_
df_test['labels'] = df_test.labels.astype('category')
```
